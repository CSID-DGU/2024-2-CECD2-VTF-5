```python
from langchain_community.chat_models import ChatOpenAI
from langchain import PromptTemplate
from langchain.chains import LLMChain
from langchain.chains import SequentialChain

llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0.5)

prompt1 = PromptTemplate(
    input_variables=['sentence'],
    template="다음 문장을 두문장 이내로 요약하세요.\n\n{sentence}"
)
chain1 = LLMChain(llm=llm, prompt=prompt1, output_key="summary")

prompt2 = PromptTemplate.from_template(
    "생애사적 자서전을 작성할 때 다음 내용으로 쓸만한 다른 주제의 질문 3개를 해주세요.\n\n{summary}"
)
chain2 = LLMChain(llm=llm, prompt=prompt2, output_key="question")

AllChain = SequentialChain(
    chains=[chain1, chain2],
    input_variables=['sentence'],
    output_variables=['summary', 'question']
)

sentence = """
하와이 주변 태평양은 세계 4대 어장의 하나임을 나는 고등학교 수산 교육 과정에서
익히 알고 있었고 이모부님도 여기 하와이에서 참치 잡으러 오셨다는 이야기도 들었
다. ‘95년 12월 31일에 나는 낚시꾼 6명을 모집하였는데 참가비는 인당 100 $였다.
아침 일찍 부두에 나가 보니 여러 낚시 배 선장들이 좋지 않은 날씨에 출항할지에 대한
의논을 하고 있었다. 우리는 한국계 3대 자손인 선장과 유럽계 선원 한 사람의 배에 탔
다. 출발 30분도 안되어 파도는 배를 요동(롤링 및 피칭)치게 했다. 금방 4명은 뻗어서
선실에 들어갔고 어부의 아들인 나와 이 공장장은 조타실에서 즐겁게 먼바다를 바라보
며 이야기를 나누었다. 낚싯대는 사람 수에 맞게 6개를 배 옆과 고물(선미)에 죽 둘러
서 배치를 했고 미끼는 고등어 모양의 가짜미끼(루어 낚시)였으며 배는 시속 11노트로
전속을 달리고 있었다. 약 2시간 정도 더 태평양 중심으로 항해했고 낚싯대는 조용하
기만 했다. 나와 공장장은 속삭였다.
“낚시는 원래 혹시나 해서 나오면 역시나 하며 돌아갑니다.”
약 3시간이나 지났을까? 갑자기 낚싯대 하나가 요동을 치며 매달아 놓은 방울 소리가
요란하게 울리며 선원은 소리쳤다. “Big Fish.” 나는 고물 Deck에 고정되어 있는 의자
에 벨트를 매고 낚싯줄을 감아올리기 시작했다. 선원이 가르쳐 주는 대로 낚싯대를 당
기고 난 다음 앞으로 숙이면서 릴을 감고 하는 작업을 반복하였다. 뒤에 계신 이 공장
장은 그때 구입했던 캠코더로 비디오 사진 찍기에 바쁘다. 이때에는 뻗어 있던 일부 다
른 사람들이 구경을 했고 약 40분 만에 모습을 드러내기 시작한 물고기는 저 멀리서부
터 좌우 지그재그로 버티며 수면 위로 오르락내리락하면서 저항하기 시작했다. 선원은
배 옆에 가까이 다가온 생선을 큰 갈고리로 후려 찍었고 우리는 갈고리에 매달린 밧줄
로 뱃전에 끌어올렸다.
"""

result = AllChain({"sentence": sentence})

print("Summary:", result["summary"])
print("Question:", result["question"])
```
